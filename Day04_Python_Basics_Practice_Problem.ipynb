```python
def main():
  # First, we need to build a list of dictionaries to hold all of the data #
  animal_sounds = [
      {"animal": "Blue Whale", "db": 188},
      {"animal": "African Lion", "db": 114},
      {"animal": "Grey Wolf", "db": 90},
      {"animal": "Honeybee", "db": 50},
      {"animal": "Domestic Cat", "db": 25}
  ]
  print("Which Animals Are Loud?")

  # Now that we have the list of animals entered, let's analyze them one by one using a 'for' loop.#
  for sound in animal_sounds:
    #Extract the animal and db from the current dictionary
    animal = sound["animal"]
    db = sound["db"]

    #Then we can use a conditional (if/else) to create a label
    if db >= 70:
      label = "loud"
    else:
      label = "quiet"

    # Print the result for each animal
    print(f"{animal}: {db} db is {label}")

# Call the main function to execute the code
if __name__ == "__main__":
  main()
```
