While allows yourself to ask a question again and again.  It takes a Boolean effect (T/F).  An alternate way of thinking about this within a loop perspective is, instead of typing "3" "3" "3"... we can take the stance that do I have any "3"s left? If while line rings true, it continues to loop. It loops between line 2 and line 3. Tip: Ctrl + C= Cancel if you happen to get accidentally stuck in an infinite loop. Below is a "countdown approach".

```python
i = 3
while i != 0:
  print("meow")
  i = i - 1
```

In order to take more of a count-up approach, you have to take the base number and then formulates so that it adds the value to the i total. (Convention is to start at number 0. The < sign indicates up to but not "through" 3.) The i = i + 1 can be shortened to i+=1. This is something special to Python.

```python
i = 0
while i < 3:
  print("Meow")
  i += 1
```

for - allows you to iterate over a list of items; list - a list of items or a variable that contains many items.  In the example below, instead of letting python choose the integer based on the occurence and then checking for matching data to print, here we are specifying within a list what our integers will be.  When i is either 0, 1, or 2, it will meow. A list is indicated by those square brackets [].

```python
for i in [0, 1, 2]:
  print("meow")
```

```python
for i in range(3):
  print("meow")
```

Let's talk about infinite loops and when it can be helpful.  This is industry standard say whenever we are looking for a definitive value.

In the example below, we know that the loop will run whenever the answer to the "while true" statment is indeed true.  We also know that we would need a positive int that is higher than zero in order for the cat to meow so we set up the parameters below, where if the input from the user is less than 0, it will continue to ask the user "What's n?" hoping for a higher value, but if for example, the input does meet the requirements (ex. 2) it would then print meow twice.

```python
while True:
  n = int(input("What's n?"))
  if n < 0:
    continue
  else:
    break
for _ in range(n):
  print("meow")
```

We can reformat this even better since we know that we need a positive int, we can wait for a positive int to be entered in order to break out of the loop constantly asking for n, and then print accordingly.

```python
while True:
  n = int(input("What's n?"))
  if n > 0:
    break
for _ in range(n):
  print("meow")
```

Putting it all together:
1.   Build out custom function- First our main one, then our custom one; Now we know that a "Meow" function does not yet exist, but we are going to build a function as if it does. Think of it like a blueprint section of our code.
2.   Now, we need to define our custom function "Meow".  We are going to say that for any number within range of n, it will print "meow".
3.  Call the main function by typing main() towards the bottom of the code.





```python
def main():
  meow(3)


def meow(n):
  for _ in range(n):
    print("meow")


main()
```

Ok great! The function is working as expected.  However, 3 is hard-coded at the top. Let's fix that. So, instead of just putting in a random number, let's make it so we ask the user to input a number.  We will need to updated the main function as well as add a section that defines that new section of custom code.

```python
def main():
  number = get_number()
  meow(number)

def get_number():
  while True:
    n = int(input("What's n?"))
    if n > 0:
      break
  return n
def meow(n):
    for _ in range(n):
      print("meow")


main()
```

TIP: In the new custom function we are working with with get_number, you can either jump straight to the return portion of code OR you can break it and then return the value of n.  BUT YOU MUST ALWAYS HAVE THAT RETURN LINE IN THERE WHETHER INSIDE OF OUTSIDE OF THE LOOP BUT ALWAYS WITHIN THE SAME FUNCTION.

Lets revisit the problem from day 1.  We want to know how many vowels are in a particular sentence.

```python
# Build a defined function for what I want to do: Count the Vowels
def count_vowels():
  vowels = "aeiou"
# Ask User for Their Sentence
  sentence=input("What's your sentence? ")
# Re-format so that the sentence is standardized.
  clean_sentence = sentence.lower()
# Count How Many Vowels are in the Sentence
  vowel_count = 0
  for char in clean_sentence:
    if char in vowels:
      vowel_count += 1
# Output to User
  return vowel_count

print(f"Total Vowels: {count_vowels()}")
```
