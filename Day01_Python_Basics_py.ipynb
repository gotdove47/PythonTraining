Let's say that during input, there is user error and they input too many spaces.  The read-out will also include those spaces unless we alter the str to account for that.

```python
# Ask user for their name
name = input("What's your name? ")

# Remove whitespace from str
name = name.strip()

# Say hello to user
print(f"hello, {name}")
```

We found the title function through the Python library that capitalizes the first letter of each word like a book title! Tip: For command terminal, you can use up arrow to select $ python name of document to quickly select

```python
# Ask user for their name
name = input("What's your name? ")

# Remove whitespace from str
name = name.strip()

# Capitalize user's name
name = name.title()

# Say hello to user
print(f"hello, {name}")
```

**Applying Methodology for AI Audio**

If this were the problem we were trying to solve: Write a script that takes a sentence, counts the number of "vowels".

The first thing I would want to do is have the user input their sentence and then based on their answer, count the number of vowels.  But we also must help the system to define what a vowel is.

```python
# Build a defined function for what I want to do: Count the Vowels
def count_vowels():
  vowels = "aeiou"
# Ask User for Their Sentence
  sentence=input("What's your sentence? ")
# Re-format so that the sentence is standardized.
  clean_sentence = sentence.lower()
# Count How Many Vowels are in the Sentence
  vowel_count = 0
  for char in clean_sentence:
    if char in vowels:
      vowel_count += 1
# Output to User
  return vowel_count

print(f"Total Vowels: {count_vowels()}")
```

Ok so here we can see that i did build a successful function, however it is not performing as expected.  I had to reformat the vowels and I did use an f-string to format the output of the results once the vowels are counted correctly.
