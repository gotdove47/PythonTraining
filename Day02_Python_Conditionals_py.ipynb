Ok this seems to be functioning properly and is correct.  However, how can I make this cleaner? You can effectively switch the operational sign to <= in the first argument and while keeping the second argument in tact for each range, nest them together. This is a unique function to Python specifically.

```python
score = int(input("Score: "))

if score 90 <= score <= 100:
  print("Grade: A")
elif score 80 <= score < 90:
  print("Grade: B")
elif score 70 <= score < 80:
  print("Grade: C")
elif score 60 <= score < 70:
  print("Grade: D")
else:
  print("Grade: F")
```

To simplify even more, we don't need to check the bounds for each range.  We know for example that anything equal to or above a 90 is going to be an "A". Parity - Asking if a number is even or odd; Before we learned that the % symbol gave us remainders. Our mission? Determine if the input is even or odd. (Hint: Even numbers won't have a remainder whilst odd numbers will. 5/2= 2 with 1 remainder; 4/2 = 2 with 0 remainders.)

```python
x = int(input("What's x? "))

if x % 2 == 0:
  print("Even")
else:
  print("Odd")
```

We don't want to recreate this again and again; so we need to define our own function to determine whether or not a number is even. First, we will just assume that the is_even function pre-exists.  (It does not, but do not worry.) We will then define what exactly the is_even function does, so that when it is called, it will work as intended.

```python
# Define main argument or process. Here we know that the if is_even function isn't quite ready yet, but we are building the shell of how the data will be read.
def main():
  x = int(input("What's x? "))
  if is_even(x):
    print("Even")
  else:
    print("Odd")
# Second, we now go down and define what this is_even function does

def is_even(n):
  if n % 2 == 0:
    return True
  else:
    return False

#Remember to call the main argument at the end for it all to work
main()
```

Tip: Python is unique again in the sense that we can condense that return statement under the is_even function into 1 line like so:

*Match conditional*

Here we are working through match conditionals based on Harry Potter and houses. Name of document? house.py. Here might be the initial set-up:

```python
name = input("What's your name? ")

if name == "Harry" or name == "Hermione" or name == "Ron":
  print("Gryffindor")
elif name == "Draco":
  print("Slytherin")
else:
  print("Who?")
```

```python
name = input("What's your name? ")

match name:
  case "Harry":
    print("Gryffindor")
  case "Hermione":
    print("Gryffindor")
  case "Ron":
    print("Gryffindor")
  case "Draco":
    print("Slytherin")
  case _:
    print("Who?")
```

Tip: The _ is used as a catch-all for anyone not mentioned. BUT there is an even better way to combine while keeping the match statement to tighten up the readability.

```python
name = input("What's your name? ")

match name:
  case "Harry"| "Hermione"| "Ron":
    print("Gryffindor")
  case "Draco":
    print("Slytherin")
  case _:
    print("Who?")
```
